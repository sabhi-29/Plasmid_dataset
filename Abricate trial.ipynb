```python
# Importing a fasta file and runing abricate on it
import os
import pandas as pd

input_file = pd.read_excel(r"My notebooks/pVAPA1637/NZ_KX443393_pVAPA1637_data.xlsx")
input_file.head()
```

```python
# Example
from Bio import Entrez, SeqIO

locus_tag = input_file['Locus Tag'][1]
Entrez.email = 'example@go.com'

search_term = f"{locus_tag}"
handle = Entrez.esearch(db = 'nucleotide', term = search_term, retmode = 'xml')
record = Entrez.read(handle)

nucleotide_id = record['IdList'][0]
print(nucleotide_id)
handle = Entrez.efetch(db = 'nucleotide', id = nucleotide_id, rettype = 'fasta', retmode = 'text')
fasta_sequence = handle.read()
# print(type(fasta_sequence))

# Saving this sequence
fasta_path = 'MG825368_fasta.fa'
with open(fasta_path, 'w') as file:
    file.write(fasta_sequence)
```

```python
# We have got the fasta file of the nucleotide sequence of the plasmid
# Now we will use abricate on this to get the resistance genes

import subprocess

input_file_fas = 'MG825368_fasta.fa'

# abricate_cmd = f'abricate {input_file_fas} --db ncbi --csv > {input_file_fas}.csv'
abricate_cmd = f'abricate {input_file_fas} --csv > {input_file_fas}_1.csv'
result = subprocess.check_output(abricate_cmd, shell = True, universal_newlines = True)


# We can see that we have obtained the desired result
```
